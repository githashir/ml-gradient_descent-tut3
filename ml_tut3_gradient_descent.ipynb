This notebook is a guide to utilizing gradient descent algorithm to optimize the loss/cost function.

We know that:

Cost Function --> MSE = L = 1/n(Sigma(Actual y - Predicted y)^2)

Linear equation --> y=mx+c   OR y = theta*x+B ; where m=theta= coefficents and B = intercept

In Gradient Descent:

m* = m-(learningrate*dL/dm)

b* = b-(learningrate*dL/db)

Our Goal is to compute the correct/efficient value of m and b that gives the minimum cost. Its is where the cost function would have converged.
